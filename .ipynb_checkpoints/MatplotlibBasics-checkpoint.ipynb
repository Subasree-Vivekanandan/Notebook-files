## Creating subplots => group of small axes in single figure

Subplot takes in 3 parameters 

1) row

2) column

3) index

rows & col specify how a snigle graph is split if (2,2) then a single graph will be split into 4 parts which can be accessed by using index

```python
x =np.arange(3)
y=2*x
plt.suptitle("Assume index starting from 1")

plt.subplot(2,2,1)
plt.title("Present at index 1")
plt.plot(x,y,'b')

plt.subplot(2,2,2)
plt.title("Present at index 2")
plt.plot(x,1-y,'r')

plt.subplot(2,2,3)
plt.title("Present at index 3")
plt.plot(x,2-y,'y')

plt.subplot(2,2,4)
plt.title("Present at index 4")
plt.plot(x,y,'g')
plt.tight_layout()
plt.show()
```

```python
fig,axs=plt.subplots(2,2,figsize=(7,7))
fig.suptitle("Assume index as a matrix(dimension is 2X2)")
axs[0,0].plot(x,y,'b')
axs[0,0].set_title("Present at index 0,0")
plt.plot()


axs[0,1].plot(x,1-y,'r')
plt.plot()
axs[0,1].set_title("Present at index 0,1")

axs[1,0].plot(x,2-y,'y--o')
plt.plot()
axs[1,0].set_title("Present at index 1,0")

axs[1,1].plot(x,y,'g--o')
plt.plot()
axs[1,1].set_title("Present at index 1,1")
plt.tight_layout()
plt.show()
```

# Linear, Exponential, Logarithmic

```python
t=np.linspace(1,10,1024)
plt.xscale('log')
plt.yscale('log')

#linear
plt.plot(t,t,label='$y=x$')
#Exponential
plt.plot(t,10**t,label='$y=10^x$')
#Logarithmic
plt.plot(t,np.log(t),label='$y=log(x)$')

plt.legend(loc=2)
plt.show()
```

# y=10**x

```python
x=np.arange(0,5)
y=10**x
plt.suptitle("Logarithmic scale")
plt.subplot(1,2,1)
plt.yscale('log')
plt.title("y=10^x")
plt.plot(x,y,label='$y=10^x$')

plt.subplot(1,2,2)
plt.xscale('log')
plt.title("x=10^y")
plt.plot(y,x,label='$x=10^y$')


plt.show()
```

# Changing base using base keyword in xscale() / yscale()

```python
x=np.arange(0,5)
y=10**x
plt.suptitle("Logarithmic scale of BASE 3")
plt.subplot(1,2,1)
plt.yscale('log',base=3)
plt.title("y=3^x")
plt.plot(x,y,'r',linewidth=3.0,label='$y=3^x$')

plt.subplot(1,2,2)
plt.xscale('log' ,base=3)
plt.title("x=3^y")
plt.plot(y,x,'b',linewidth=3.0,label='$x=3^y$')


plt.show()
```

## Sinusoidal wave + Cosine wave

```python
import math
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0,2.10,0.10)
y1=np.sin(math.pi*t)
y2=np.cos(math.pi*t)
y3=np.cos(math.pi*t+math.pi/2)
plt.grid(True)
plt.plot(t,y1,'b',label="sine wave",linewidth="3.0") # linewidth for thickness
plt.plot(t,y2,'r',label="cosine wave")
plt.plot(t,y3,'y',label="cosine wave with phase shift")
plt.legend(bbox_to_anchor=(1.02,1.0))

plt.show()
```

```python
time=np.arange(0,2.10,0.1)
sint=np.sin(math.pi*t)
cost=np.cos(math.pi*t)
sinets=np.sin(math.pi*t+math.pi/2)
costs=np.cos(math.pi*t+math.pi/2)\

fig,ax=plt.subplots(2,2,figsize=(6,6))
ax[0,0].plot(time,sint,'--b')
plt.plot()
ax[0,0].set_title("sine")

ax[0,1].plot(time,cost,'-.r')
plt.plot()
ax[0,1].set_title("cosine")

ax[1,0].plot(time,sinets,':y')
plt.plot()
ax[1,0].set_title("sine shift")

ax[1,1].plot(time,costs,'-g')
plt.plot()
ax[1,1].set_title("cosine shift")

plt.tight_layout()
plt.plot()
plt.show()
```

# Adding Annotations

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,10)
y=x
y1=10-x

fig,ax=plt.subplots()

plt.plot(x,y,label="Demand")
plt.plot(x,y1,label="Supply")

ax.annotate("Equilibrium",xy=(5,5),xytext=(4,2), \
            fontsize=10,fontweight='semibold', \
            arrowprops=dict(arrowstyle="->",linewidth=2)
)

plt.xlabel("quantity")
plt.ylabel("price")

plt.legend()
plt.show()
```

```python
from matplotlib.patches import Circle,Polygon
from matplotlib.collections import PatchCollection

fig,ax=plt.subplots()
patches=[]

circle=Circle((.4,.3),.13)
triangle=Polygon([[.2,.5],[.3,.7],[.3,.5]],True)

patches+=[circle,triangle]

colors =100*np.random.rand(len(patches))
p=PatchCollection(patches)
p.set_array(np.array(colors))
ax.add_collection(p)

plt.show()
```

# User defined Universal function

### numpy.frompyfunc(func, nin, nout)

Parameters: 


func: [A python function object ] An arbitrary python function


nin: [int] Number of input arguments to that function. 


nout: [int] Number of objects returned by that function.


Return: A Numpy universal function object.

```python
x=np.arange(1,20)

def modulo(val):
    return val%10

mod10=np.frompyfunc(modulo,1,1)

modres=mod10(x)
print("Modulo by 10 :",*modres)
```

# M=Q*R

M-> square/rectangular matrix
Q-> orthogonal matrix
R-> upper triangular matrix

```python
x=np.mat("3 1 4;1 5 7;2 6 5")
```

```python
b=np.array([1,2,3]).reshape(3,1)
b
```

```python
q,r=np.linalg.qr(x)
res=np.dot(np.linalg.inv(r),np.dot(q.T,b))
print(res)
```

```python
#Check if its true
res= np.linalg.solve(x,b)
res
```

## Reshaping

```python
ath = oo[(oo.Year==2008)&((oo.Event=="100m")|(oo.Event=="200m"))]
ath
```

```python
g =ath.groupby(["Discipline","Event","Country_Code","Gender"]).size()
g
```

```python
df = g.unstack(["Discipline","Event"])
df
```

### stack()

Create a taller dataframe/series

```python
df.stack()
```

```python
df.stack("Event")
```

### unstack()

Create a wider dataframe/series

```python
df.unstack()
```

```python
df.unstack("Gender")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
X = df.iloc[:,[3,4]].values
X
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init="k-means++",random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title("The elbow method")
plt.xlabel("No of Clusters")
plt.ylabel("WCSS(within cluster sum of squares)")
plt.show();
```

## Training  K-Means model on the datset

```python
kmeans = KMeans(n_clusters=5, init="k-means++",random_state=42)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.fit_predict(X)  # fir_predict returns value for dependent variable i.e. value for 5 clusters
```

## Visualising the clusters

```python
plt.scatter(X[y_kmeans==0, 0],X[y_kmeans== 0,1],s=100,color="red",label="Cluster1")
plt.scatter(X[y_kmeans==1, 0],X[y_kmeans== 1,1],s=100,color="blue",label="Cluster2")
plt.scatter(X[y_kmeans==2, 0],X[y_kmeans== 2,1],s=100,color="green",label="Cluster3")
plt.scatter(X[y_kmeans==3, 0],X[y_kmeans== 3,1],s=100,color="cyan",label="Cluster4")
plt.scatter(X[y_kmeans==4, 0],X[y_kmeans== 4,1],s=100,color="magenta",label="Cluster5")

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='y',label="centroids")
plt.xlabel("Annual Income(k$)")
plt.ylabel("Spending score (1-100)")
plt.title("Clusters of customers")
plt.legend();
plt.show();
```

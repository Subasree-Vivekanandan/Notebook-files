```python
# Subplots using different axes => ax1,ax2,ax3,ax4

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2)        # we refer ax1,ax2,ax3,ax4, in a tuple
x1=np.array([1,2,3,4,5])
ax1.plot(x1,x1**2)
ax1.set(title="Subplot 1",
       xlabel="x1",
       ylabel="x1^^2")
ax2.scatter(x,np.sin(x),color="pink")
ax2.set(title="Subplot 2",
       xlabel="x",
       ylabel="sin x")
ax3.bar(batsman,runs,color="green")
ax3.set(title="Subplot 3",
       xlabel="batsman",
       ylabel="runs")
ax4.hist(np.random.randn(1000),color="aqua");
ax4.set(title="Subplot 4",
       xlabel="1000 random samples",
       ylabel="Normal distribution")

fig.tight_layout()   # tight_layout to create proper space between subplots

fig.savefig(r"C:\Users\HP\Notebook files\DA - pandas,numpy,scikit learn\plt generated\subplots.png")
```

```python
fig,(ax1,ax2)= plt.subplots(nrows=2,ncols=1,figsize=(10,5))

heart_df["chol_mean"].plot(linestyle="dashed",xlim=(0,100),ax=ax1,color="red")
heart_df.plot(ax=ax1,x="age",y="chol",kind="scatter");
ax1.legend()

heart_df["chol_mean"].plot(linestyle="dashed",xlim=(0,100),ax=ax2,color="red")
heart_df.plot(ax=ax2,x="age",y="chol",kind="scatter");
ax2.legend()

fig.tight_layout()
```

```python
# scatter plotting of cholestestrol level of people who's age is greater than 50

over_50 = heart_df[heart_df["age"]  > 50]
over_50.head()
```

```python
over_50["target"].unique()  # target specifies 1 for people with heart disease , 0 for people not having heart disease
```

```python
fig,ax = plt.subplots(figsize=(10,5))

over_50.plot(x="age",y="chol",
             kind="scatter",
             c="target",
             ax=ax,
              );
```

```python
fig,(ax,ax1) = plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,5));

over_50.plot(x="age",y="chol",
             kind="scatter",
             c="target",
             ax=ax);

over_50.plot(x="age",
            y="chol",
            kind="scatter",
            color="red",
            ax=ax1);

fig.tight_layout()
```

```python
plt.style.use("seaborn-whitegrid")
# Actual OO method

fig,(ax1,ax2)= plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,6))

plot1 = ax1.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter"
                   );
ax1.set(xlabel="age",
        ylabel="cholesterol",
       title="Cholesterol vs Age > 50");

ax1.set_xlim([50,80])

ax1.axhline(y=over_50["chol"].mean(),linestyle="--");

ax1.legend(*plot1.legend_elements(),title="target");


plot2= ax2.scatter(x=over_50["age"],
                  y=over_50["thalach"],
                  c=over_50["target"],
                  cmap="summer");
ax2.set(xlabel="age",
       ylabel="max heart rate",
       title="Age Vs Max heart rate");

ax2.set_xlim([50,80])
ax2.set_ylim([60,200])
ax2.axhline(y=over_50["thalach"].mean(),linestyle="--");

ax2.legend(*plot2.legend_elements(),title="target");

fig.suptitle("Heart disease analysis",fontsize=14,fontweight="bold")
fig.tight_layout()


fig.savefig(r"C:\Users\HP\Notebook files\DA - pandas,numpy,scikit learn\plt generated\subplots-scatter.png")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"Udacity_final.csv",encoding='mac_roman')
df.head()
```

```python
df.shape
```

```python
list_null = df.isna().sum()
```

```python
list(list_null)
```

```python
x = list(df.columns)
x
```

```python
y = list(list_null)
```

```python
import seaborn as sns


def addlabel(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha = 'center',bbox = dict(facecolor = 'yellow', alpha =.8))
        
        
fig,ax = plt.subplots(figsize=(7,6))
sns.barplot(x,y,ax=ax)
ax.set_title("Null value distribution in columns",fontdict={"fontweight":"bold","fontsize":15},pad=15.0)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
ax.set_ylabel("Count of Rows with null values",fontdict={"fontweight":"bold","fontsize":12})
ax.set_xlabel("Organization",fontdict={"fontweight":"bold","fontsize":12})
addlabel(x,y)

ax.figure
```
